```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Load and merge PLUTO data from each borough
bk = pd.read_csv(r'..\Data\pluto\BK.csv')
bx = pd.read_csv(r'..\Data\pluto\BX.csv')
mn = pd.read_csv(r'..\Data\pluto\MN.csv')
qn = pd.read_csv(r'..\Data\pluto\QN.csv')
si = pd.read_csv(r'..\Data\pluto\SI.csv')

pluto = pd.concat([bk, bx, mn, qn, si])

pluto.reset_index(inplace=True)
pluto = pluto.drop('index',axis=1)
pluto.head()
```

```python
# Load LL84 data for year 2015
ll84 = pd.read_csv(r'..\Data\LL84_2015.csv')

# Select property types of interest
ll84 = ll84[(ll84['Primary Property Type - Self Selected'] == 'Office')|
            (ll84['Primary Property Type - Self Selected'] == 'Multifamily Housing')|
           (ll84['Primary Property Type - Self Selected'] == 'Hotel')]

ll84.reset_index(inplace=True)
ll84 = ll84.drop('index',axis=1)
ll84.head()
```

Notice that PLUTO and LL84 BBL values are not reported the same way. In LL84 are reported as strings (often including dashes), whereas in PLUTO as floats. The value types need to be consistent to be able to merge.

```python
ll84['NYC Borough, Block and Lot (BBL)'].replace('-','',regex=True,inplace=True)
ll84['NYC Borough, Block and Lot (BBL)'] = pd.to_numeric(ll84['NYC Borough, Block and Lot (BBL)'], errors = coerce)
```

```python
# Merge files
df = ll84.merge(pluto, left_on='NYC Borough, Block and Lot (BBL)', right_on='BBL')
df.to_csv('..\Data\LL84_PLUTO.csv')
```

After merging, notice that we are losing 71 properties. This could be attributed either in inconsistancies between buildings reported in LL84 and PLUTO, or other types of BBL formatting that were not able to capture. Nevertheless, the loss in data is negligible compared to the size of the dataset.
